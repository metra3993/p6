Цель: в данной практической работе необходимо продемонстрировать датасет с практической работы 5 в виде графиков, диаграмм и так далее. 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas  as pd
greek_gods = pd.read_csv(r"C:\Users\user\Downloads\greek_gods2.csv")
greek_gods
```

Мы импортировали наш датасет о нетлфиксе при помощи read_csv 

```python
plt.figure(figsize=(20,10))
plt.plot(greek_gods['name-greek'].tail(5), greek_gods['description'].tail(5), label = 'Зависимость')
plt.title('Зависимость имени от описания')
plt.xlabel('Имя')
plt.ylabel('Описание')
plt.legend()
plt.show()
```

Создаем новую график с заданным размером. Строим график линии, где по оси x отображаются последние 5 значений столбца 'name-greek', а по оси y - последние 5 значений столбца 'description'. label используется для создания метки легенды. Устанавливаем заголовок для графика. Устанавливаем подписи осей x и y соответственно. Добавляем легенду, которая в данном случае будет отображать метку 'Зависимость'.
Таким образом, этот код создает линейный график, отражающий статистику зависимости имени Бога и его прелназначения в нашем датасете.

```python
plt.figure(figsize=(10, 5))
plt.plot(greek_gods['name-english'].head(7), greek_gods['name-greek'].head(7), label='Имена', color = "red")
plt.bar(greek_gods['name-english'].head(7), greek_gods['name-greek'].head(7), label='Имена (столбцы)', color = "pink")
plt.title('Сопоставление греческого имени с английским')
plt.xlabel('Имя на английском')
plt.ylabel('Имя на греческом')
plt.legend()
plt.show()
```

На данном графике показано наложение линейного графика на столбчатую диаграмму. Показано сопоставление английского имени бога с греческим

```python
plt.plot (greek_gods["name-english"].tail(7), greek_gods["main-type"].tail(7), color = "green", marker = "o", label = "Основной тип")
plt.plot (greek_gods["name-english"].tail(7), greek_gods["sub-type"].tail(7), color = "blue", marker = "o", label = "Подтип")
plt.legend()
plt.show()
```

В данном графике показано наложение двух линейных графиков со значениями типов и подтипов. Видно что все имена присущи Богам, но подтипы у них разные.

```python
plt.bar(greek_gods["name-english"].tail(7), greek_gods["sub-type"].tail(7), color = "purple")
plt.show()
```

На данной столбчатой диаграмме показана зависимость имени греческого Бога на английском языке от его подтипа.

```python
plt.bar(greek_gods["name-english"].tail(7), greek_gods["sub-type"].tail(7), color = "blue")
plt.bar(greek_gods["name-english"].tail(7), greek_gods["main-type"].tail(7), color = "green")
plt.show()
```

На данной столбчатой диаграмме показано наложение двух столбчатых диаграмм со значениями типов и подтипов. Должно быть видно (но не видно, потому что все значения типа обжект) что все имена присущи Богам, но подтипы у них разные.

```python
data = greek_gods["sub-type"].tail(30).value_counts()
plt.pie(data, labels = data.index, autopct = "%1.1f%%")

plt.show()
```

На данной круговой диаграмме видно, что подтипы нимфа, муза и время состовляют 29.9%, а остальные проценты составляют другие подтипы.

```python
timegame = greek_gods['description'].tail(20).value_counts()
valid_keys = timegame[timegame.index]
labels = valid_keys.index
plt.pie(valid_keys, labels=labels, autopct='%1.1f%%', colors = ['green','violet','lightblue','yellow'])
plt.title("Соотношение предназначений Богов")
plt.show()
```

Данная круговая диаграмма показывает соотношение предназначений Богов. По 10% имеют призвания быть Местной богиней плодородия и сыном Геракла, руководящий обороной укрепленных городов и цитаделей. 15% имеет призвание быть дочерью Аполлона. Все остальные призвания имею по 5%.

```python
plt.figure(figsize=(10,7))
plt.scatter(greek_gods['name-english'].tail(10), greek_gods['sub-type'].tail(10), color = 'red')
plt.title('Категории богов')
plt.xlabel('Имена')
plt.ylabel('подтипами')
plt.show()
```

На данной диаграмме рассеивания показана связь между английскими именами богов и их подтпами. Так же на диаграмме отсутсвует рассеивание.

```python
plt.figure(figsize=(10,7))
plt.scatter(greek_gods['name-greek'].head(15), greek_gods['sub-type'].head(15), color = 'red')
plt.title('Происхождение Богов')
plt.xlabel('Боги')
plt.ylabel('Подтипы')
plt.show()
```

На данной диаграмме рассеивания показано, что первые 15 Богов произошли на Олимпе, а у последнего Бога исконное происхождение. Так же на диаграмме отсутсвует рассеивание.

```python
plt.figure(figsize=(20,10))
plt.hist(greek_gods['description'].head(40), bins=30, color = "green", edgecolor = 'blue', alpha=0.6, label = 'Предназначения')
plt.legend()
plt.show()
```

В данной гисторгамме показаны предназначения богов(датасет значений int не имеет, поэтому графики и обьяснения такие)

```python
plt.figure(figsize=(20,10))
plt.hist(greek_gods['sub-type'].tail(60), bins=30, color = "black", edgecolor = 'red', alpha=0.6, label = 'Подтипы')
plt.legend()
plt.show()
```

В данной гисторгамме показаны подтипы Богов.

```python
plt.figure(figsize=(40,10))
plt.hist(greek_gods['description'].tail(60), bins=30, color = "blue", edgecolor = 'orange', alpha=0.6, label = 'Предназначения')
plt.legend()
plt.show()
```

В данной гисторгамме показаны описания предназначений богов. В среднем все значения повторяются по 2 раза, но значение 1% повторяется только 4 раза, а значение 4% повторяется 2 раза.
